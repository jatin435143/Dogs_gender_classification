```python
columns=data.columns
```

### DIVIDING THE COLUMNS INTO CATEGORICAL,NUMERICAL AND TARGET ONES

```python
categorical=[]
numerical=[]
target=data.columns[-1]
for column in columns[0:-1]:
    if(type(data[column][0])==str):
        categorical.append(column)
    else:
        numerical.append(column)
```

```python
categorical,numerical,target
```

### CONVERTING THE CATEGORICL DATA INTO NUMERICAL USING ONE HOT ENCODING

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the one-hot encoder
one_hot_encoder = OneHotEncoder()

# Apply one-hot encoding to the 'Color' column
one_hot_encoded = one_hot_encoder.fit_transform(data[categorical])

# Convert the result to a DataFrame
data_one_hot_sklearn = pd.DataFrame(one_hot_encoded.toarray(), columns=one_hot_encoder.get_feature_names_out(categorical))

# Concatenate the new one-hot encoded columns to the original dataframe
data = pd.concat([data, data_one_hot_sklearn], axis=1)
```

```python
data.columns
```

```python
data=data.drop(columns=categorical) #AS NOW HAVE NUMERICAL COLS OF CATEGORICAL ONES SO NO NEED OF THEM
```

```python
cols_need=data.columns.drop(target) # THE COLS OTHER THAN TARGET
```

### CONVERT EACH AND EVERY ELEMENT TO SMALL NUMBER TO REDUCE BIASNESS

```python
from sklearn.preprocessing import StandardScaler
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
scaled_data = scaler.fit_transform(data[cols_need])
data_scaled = pd.DataFrame(scaled_data, columns=cols_need)
```

```python
data_scaled
```

### BUILDING KNN MODEL FROM SCRATCH

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(data_scaled,data[target],test_size=0.2)
X_train_np,X_test_np,Y_train_np,Y_test_np=X_train.to_numpy(),X_test.to_numpy(),Y_train.to_numpy(),Y_test.to_numpy()
```

```python
class Knn:
    def __init__(self,k):
        self.k=k

    def fit(self,x,y):
        self.x=x
        self.y=y
        
    def eucledian(self,x1,x2):
        return np.sqrt(np.sum((x1-x2)**2,axis=1))
        
    def _predict(self,single): # for single dataset
        predictions=self.eucledian(self.x,single)
        ppop=np.argsort(predictions)[:self.k]
        real_predict=np.array(self.y[ppop])
        return real_predict
        
    def predict(self,x1):
        predictions=[self._predict(np.array(element)) for element in x1]
        real_predicts=[]
        for i in predictions:
            unique_values, counts = np.unique(i, return_counts=True)
            max_count = np.argmax(counts)
            max_value = unique_values[np.argmax(counts)]
            real_predicts.append(max_value)
        return real_predicts
        
    def accuracy_classification(self,y_predict,y1):
        ppoo=np.array(y_predict==y1)
        acc=(np.sum(ppoo)/len(y1))*100
        return acc
        
```

```python
array=[]
for k in range(1,40,2):
    clf=Knn(k)
    clf.fit(X_train_np,Y_train_np)
    predictions=clf.predict(X_test_np)
    array.append(clf.accuracy_classification(Y_test_np,predictions))
```
